### Step 3.5: Handling Structural Missing and Duplicate Values

Chandigarh has no rural population as per PLFS survey design.
Therefore, rows corresponding to Chandigarh (Rural sector) are removed,
as they are not applicable for analysis.

```python
df_original = df.copy()
```

```python
# Remove missing values 
df_cleaned = df.dropna(subset=['value']).copy()
```

```python
# Checking the dataset shape for reducing any error
print("Original shape:", df.shape)
print("Cleaned shape:", df_cleaned.shape)
```

```python
# checking for duplicate rows
print("Duplicate rows after cleaning:", df_cleaned.duplicated().sum())
```

## Step 4: Rename Indicators and verify year

- Renaming indicator abbreviations to descriptive names for better understanding.
- Confirming this is single year dataset (2023-24).

```python
# Renaming indicator names 
df_cleaned['indicator'] = df_cleaned['indicator'].replace({
    'UR': 'Unemployment Rate',
    'WPR': 'Worker Population Ratio',
    'LFPR': 'Labour Force Participation Rate'
})
```

```python
# Confirming only one survey year exist
print(f"Year in dataset: {df_cleaned['year'].unique()}")
```

## Step 5: Split into State and National Datasets

The dataset contains two types of records 
- State - Data for individual state/UTs
- National - All India aggregate data(all India row )

The split is necessary to make datasets reliable. 
If both results are kept together and compute averages then all India row get mixed in state leads to wrong outcomes.


```python
# Split datasets into state and national datasets
df_state = df_cleaned[df_cleaned['level'] == 'State'].copy()
df_india = df_cleaned[df_cleaned['level'] == 'National'].copy()
```

```python
# Drop 'level' column after splitting
df_state = df_state.drop(columns=['level'])
df_india = df_india.drop(columns=['level'])
```

```python
# Keep only relevant columns
cols = ['year', 'state', 'sector', 'gender', 'indicator', 'value']

df_state = df_state[cols]
df_india = df_india[cols]
```

```python
# Preview national dataset
print("NATIONAL DATASET:")
df_india
```

## Step 6: Data Validation Checks (Post-Cleaning)

Performing final validation checks to ensure:
- Data consistency after cleaning
- No unintended data loss
- Structural completeness of dataset

```python
# Validating dataset structure by checking total records and unique categories 
# (states, indicators, and sectors) after cleaning
print("Final cleaned dataset shape:", df_cleaned.shape)
print("Unique states:", df_cleaned['state'].nunique())
print("Unique indicators:", df_cleaned['indicator'].unique())
print("Unique sectors:", df_cleaned['sector'].unique())
```

```python
# Validating that all combinations of indicator, gender, and sector are present 
# and have consistent record counts
print("Check combinations:")
print(df_cleaned.groupby(['indicator', 'gender', 'sector']).size().head(10))
```

```python
# Sorting the dataset by state, indicator, gender, and sector to maintain a consistent and logical order
df_cleaned = df_cleaned.sort_values(by=['state', 'indicator', 'gender', 'sector'])
```

```python
# Resetting index to ensure a clean and continuous row numbering after data cleaning
df_cleaned = df_cleaned.reset_index(drop=True)
```

## Step 7: Export Cleaned Dataset 

```python
# Exported clean dataset
df_cleaned.to_csv("../data/processed/cleaned.csv", index=False)

# State dataset
df_state.to_csv("../data/processed/state_data.csv", index=False)
print("State Data CSV Saved Successfully")
 
# National Dataset
df_india.to_csv("../data/processed/national_data.csv", index=False)
print("National Data CSV Saved Successfully")
```
