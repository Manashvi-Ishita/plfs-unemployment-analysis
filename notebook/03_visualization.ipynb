# Unemployment Analysis - Visualization

This notebook focuses on visualization key insights derived from cleaned dataset.

### Objectives:
- Compare unemployment across states
- Analyze gender gap
- Compare rural vs urban trends
- Understand labour force participation

```python
# Import the important libraries for visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df_state= pd.read_csv("../data/state_data.csv")
df_national = pd.read_csv("../data/national_data.csv")
```

## Top 10 States by Unemployment Rate

This chart shows the states with highest unemployment rates, along with national benchmark for comparison.

```python

# Filter top 10 states
top_states = df_state[
    (df_state['indicator'] == 'Unemployment Rate') &
    (df_state['sector'] == 'Rural+Urban') &
    (df_state['gender'] == 'Person')
].sort_values(by='value', ascending=False).head(10)
```

```python
# Plotting the graph for unemployment rate
plt.figure(figsize=(10,6))

top_states = top_states.sort_values(by='value', ascending=True)

plt.barh(top_states['state'], top_states['value'])

plt.title("Top 10 States by Unemployment Rate")
plt.xlabel("Unemployment Rate")
plt.ylabel("State")

plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("../images/top_states.png", bbox_inches='tight')
plt.show()
```

##  Gender Gap in Unemployment

This chart compares unemployment rates between males and females to identify gender-based disparities in employment.

```python
# Filter data for gender comparison
gender_data = df_state[
    (df_state['indicator'] == 'Unemployment Rate') &
    (df_state['sector'] == 'Rural+Urban') &
    (df_state['gender'].isin(['Male', 'Female']))
]
```

```python
# Calculate average unemployment
gender_avg = gender_data.groupby('gender')['value'].mean()

# Plot
plt.figure(figsize=(6,5))

plt.bar(gender_avg.index, gender_avg.values)

plt.title("Gender Gap in Unemployment")
plt.xlabel("Gender")
plt.ylabel("Unemployment Rate")

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("../images/gender_gap.png", bbox_inches='tight')
plt.show()
```

##  Rural vs Urban Unemployment

This chart compares unemployment rates between rural and urban areas to understand structural differences in employment patterns.

```python
# Filter rural and urban data
sector_data = df_state[
    (df_state['indicator'] == 'Unemployment Rate') &
    (df_state['gender'] == 'Person') &
    (df_state['sector'].isin(['Rural', 'Urban']))
]
```

```python
# Calculate average
sector_avg = sector_data.groupby('sector')['value'].mean()

# Plot
plt.figure(figsize=(6,5))

plt.bar(sector_avg.index, sector_avg.values)

plt.title("Rural vs Urban Unemployment")
plt.xlabel("Sector")
plt.ylabel("Unemployment Rate")

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("../images/rural_urban.png", bbox_inches='tight')
plt.show()
```

##  Labour Force Participation Rate (LFPR)

This chart compares workforce participation between males and females, highlighting differences in labour market engagement.

```python
# Filter LFPR data
lfpr_data = df_state[
    (df_state['indicator'] == 'Labour Force Participation Rate') &
    (df_state['sector'] == 'Rural+Urban') &
    (df_state['gender'].isin(['Male', 'Female']))
]
```

```python
# Calculate average LFPR
lfpr_avg = lfpr_data.groupby('gender')['value'].mean()

# Plot
plt.figure(figsize=(6,5))

plt.bar(lfpr_avg.index, lfpr_avg.values)

plt.title("LFPR by Gender")
plt.xlabel("Gender")
plt.ylabel("LFPR")

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("../images/lfpr.png", bbox_inches='tight')
plt.show()
```
